        *Brand and price: explore weather cars with the brand nissan tend to have the highest prices 

```python
data=df
df['brand'] = df['model'].apply(lambda x: x.split()[0])

# Calculate the average price for each brand
average_prices_by_brand = df.groupby('brand')['price'].mean().sort_values(ascending=False)

# Check if Nissan has the highest average price
highest_price_brand = average_prices_by_brand.idxmax()
highest_price = average_prices_by_brand.max()

# Output the results
print("Average prices by brand:")
print(average_prices_by_brand)

if highest_price_brand == 'nissan':
    print(f"The brand with the highest average price is Nissan with an average price of ${highest_price:.2f}.")
else:
    print(f"The brand with the highest average price is {highest_price_brand} with an average price of ${highest_price:.2f}, not Nissan.")
```
